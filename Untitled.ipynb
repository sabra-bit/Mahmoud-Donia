```python
import pandas as pd
df = pd.read_excel('knowlage.xlsx')  # Read the entire file as a DataFrame
```

```python
df1= df[['Activity','Activity ID Fixed Part']]
df2= df[['Activity','Activity ID Fixed Part','Successor (1) Fixed ID','Original Duration (Default Value)']]

merged_df = pd.merge(df1, df2, left_on='Activity ID Fixed Part', right_on='Successor (1) Fixed ID', how='inner')
merged_df
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

def cosine_similarity(v1, v2):
  dot_product = np.dot(v1, v2)
  norm_v1 = np.linalg.norm(v1)
  norm_v2 = np.linalg.norm(v2)
  return dot_product / (norm_v1 * norm_v2)

def compare_text_cosine_similarity(text1, text2):
  vectorizer = CountVectorizer()
  vectors = vectorizer.fit_transform([text1, text2])
  vector1, vector2 = vectors.toarray()
  similarity = cosine_similarity(vector1, vector2)
  return similarity

# Example usage
text1 = "Gravel layer works"
text2 = "gravel works"
similarity = compare_text_cosine_similarity(text1, text2)
print("Cosine similarity:", similarity)
```

```python
addDF = pd.DataFrame()
print(f" input text: {text2}")
for index, row in df.iterrows():
  similarity = compare_text_cosine_similarity(text2, row['Activity'])
  if similarity >= 0.70:
      new_row = {'Activity':row['Activity'],'similarity':similarity,'Duration':row['Original Duration (Default Value)']}
      addDF = addDF.append(new_row, ignore_index=True)
      print(f"Similarity with row {index}: {similarity} word :{row['Activity']}")

addDF = addDF.sort_values(by='similarity', ascending=False).iloc[0]
addDF
```
